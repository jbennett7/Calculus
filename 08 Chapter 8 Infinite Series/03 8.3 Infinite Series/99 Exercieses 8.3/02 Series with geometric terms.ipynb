# Series with Geometric Terms
In Exercies 7&#151;14, write out the first few terms of each series to show how the series starts. Then find the sum of the series.

## 7

$
\begin{align}
\sum^{\infty}_{n=0}{\frac{(-1)^n}{4^n}}
\end{align}
$

### Solution

$
\begin{align}
\sum^{\infty}_{n=0}{\frac{(-1)^n}{4^n}} = 1 - \frac{1}{4} + \frac{1}{16} - \frac{1}{64} + \cdots
\end{align}
$

$
\begin{align}
\sum^{\infty}_{n=0}{\frac{(-1)^n}{4^n}} = \sum^{\infty}_{n=1}{\frac{(-1)^{n-1}}{4^{n-1}}}
= \sum^{\infty}_{n=1}{\underbrace{(1)}_{a} \underbrace{\left(-\frac{1}{4^{n-1}}\right)}_{r}}
\end{align}
$

$
\begin{align}
\frac{1}{1+(1/4)} = \frac{1}{(5/4)} = \frac{4}{5}
\end{align}
$


## 8

$
\begin{align}
\sum^{\infty}_{n=2}{\frac{1}{4^n}}
\end{align}
$

### Solution

$
\begin{align}
\sum^{\infty}_{n=2}{\frac{1}{4^n}} = 1 + \frac{1}{4} + \frac{1}{16} + \frac{1}{64} + \cdots
\end{align}
$

$
\begin{align}
\sum^{\infty}_{n=2}{\frac{1}{4^n}} = \sum^{\infty}_{n=1}{\frac{1}{4^{n+1}}} 
= \sum^{\infty}_{n=1}{\underbrace{\left(\frac{1}{16}\right)}_{a}\underbrace{\left(\frac{1}{4^{n-1}}\right)}_{r}}
\end{align}
$

$
\begin{align}
\frac{1/16}{1-(1/4)} = \frac{1/16}{(3/4)} = \left(\frac{1}{16}\right)\left(\frac{4}{3}\right) = \frac{1}{12}
\end{align}
$

## 9

$
\begin{align}
\sum^{\infty}_{n=1}{\frac{7}{4^n}}
\end{align}
$

### Solution

$
\begin{align}
\sum^{\infty}_{n=1}{\frac{7}{4^n}} = 7 + \frac{7}{4} + \frac{7}{16} + \frac{7}{64} + \cdots
\end{align}
$

$
\begin{align}
\sum^{\infty}_{n=1}{\frac{7}{4^n}}
= \sum^{\infty}_{n=1}{\underbrace{\left(\frac{7}{4}\right)}_{a}\underbrace{\left(\frac{1}{4^{n-1}}\right)}_{r}}
\end{align}
$

$
\begin{align}
\frac{7/4}{1-(1/4)} = \left(\frac{7}{4}\right)\left(\frac{4}{3}\right) = \frac{7}{3}
\end{align}
$

## 10

$
\begin{align}
\sum^{\infty}_{n=0}{(-1)^n\left(\frac{5}{4^n}\right)}
\end{align}
$

### Solution

$
\begin{align}
\sum^{\infty}_{n=0}{(-1)^n\left(\frac{5}{4^n}\right)} = 5 - \frac{5}{4} + \frac{5}{16} + \frac{5}{64} + \cdots
\end{align}
$

$
\begin{align}
\sum^{\infty}_{n=0}{(-1)^{n}\left(\frac{5}{4^n}\right)} =
\sum^{\infty}_{n=1}{\underbrace{(5)}_{a}\underbrace{{\left(\frac{-1}{4}\right)}^{n-1}}_{r}}
\end{align}
$

$
\begin{align}
\frac{5}{1+(1/4)} = \left(\frac{5}{1}\right)\left(\frac{4}{5}\right) = 4
\end{align}
$
