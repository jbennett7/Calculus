# Geometric series
__Geometric series__ are series of the form:

$$\label{eq1}\tag{1}
a + ar + ar^2 + \cdots + ar^{n-1} + \cdots = \sum^{\infty}_{n=1}{ar^{n-1}}
$$

in which $a$ and $r$ are fixed real numbers and $a \ne 0$. The __ratio__ $r$ can be positive, as in

$$
1 + \frac{1}{2} + \frac{1}{4} + \cdots + \left(\frac{1}{2}\right)^{n-1} + \cdots\text{,}
$$

or negative, as in

$$
1 - \frac{1}{3} + \frac{1}{9} - \cdots + \left(-\frac{1}{3}\right)^{n-1} + \cdots\text{.}
$$

If $r = 1$, the $n$th partial sum of the series in $\eqref{eq1}$ is

$$
s_n = a + a(1) + a(1)^2 + \cdots + a(1)^{n-1} = na\text{,}
$$

and the series diverges because $\lim_{n\rightarrow\infty} = \pm\infty$, depending on the sign of $a$. If $r=-1$, the series diverges because the $n$th partial sums alternate between $a$ and 0. If $|r| \ne 1$, we can determine the convergence or divergence of the series in the following way:

\begin{align}
s_n  &= a + ar + ar^2 + \cdots + ar^{n-1} \\
rs_n &= ar + ar^2 + \cdots + ar^{n-1} + ar^n \\
s_n - rs_n &= a - ar^n \\
s_n(1 - r) &= a(1 - r^n) \\
s_n &= \frac{a(1 - r^n)}{1 - r}\text{,}\ \ \ \ \ \ \ (r \ne 1)\text{.}
\end{align}

If $|r| \lt 1$, then $r^n \rightarrow 0$ as $n \rightarrow \infty$ (as in [Section 8.2](..)) and $s_n \rightarrow a/(1 - r)$. If $|r| \gt 1$, then $|r^n| \rightarrow \infty$ and the series diverges.

If $|r| \lt 1$, the geometric series $a + ar + ar^2 + \cdots + ar^{n-1} + \cdots$ converges to $a/(1-r)$:

$$\label{eq2}\tag{2}
\sum^{\infty}_{n=1}{ar^{n-1}} = \frac{a}{1 - r},\ \ \ \ \ |r| \lt 1\text{.}
$$

__EXAMPLE 1__    The geometric series with $a = 1/9$ and $r = 1/3$ is

$$
\frac{1}{9} + \frac{1}{27} + \frac{1}{81} + \cdots = \sum^{\infty}_{n=1}{\frac{1}{9}\left(\frac{1}{3}\right)^{n-1}} = \frac{1/9}{1 - (1/3)} = \frac{1}{6}\text{.}
$$

__EXAMPLE 2__. The series

$$
\sum^{\infty}_{n=1}{\frac{(-1)^n5}{4^n}} = -\frac{5}{4} + \frac{5}{16} - \frac{5}{64} + \cdots
$$

is a geometric series with $a = -5/4$ and $r = -1/4$. It converges to

$$
\frac{a}{1 - r} = \frac{-5/4}{1+(1/4)} = -1\text{.}
$$

__EXAMPLE 3__ You drop a ball from $a$ meters above a flat surface. Each time the ball hits the surface after falling a distance $h$, it rebounds a distance $rh$, where $r$ is positive but less than $1$. Find the total distance the ball travels up and down (Fig 8.10).

__Solution__ The total distance is

\begin{align}
s = a + \underbrace{2ar + 2ar^2 + 2ar^3 + \cdots}_{\text{This sum is }2ar/(1-r)\text{.}} = a + \frac{2ar}{1 - r} = a \frac{1 + r}{1 - r}\text{.}
\end{align}

If $a = 6$m and $r = 2/3$, for instance, the distance is

$$
s = 6 \frac{1+(2/3)}{1-(2/3)} = 6\left(\frac{5/3}{1/3}\right) = 30\text{m.}
$$

__EXAMPLE 4__

Express the repeating decimal $5.23\ \ 23\ \ 23\ \ \ldots$ as the ratio of two integers.

__Solution__

$
\begin{align}
a &= 1 \text{,}\\
r &= 1/100
\end{align}
$

\begin{align}
5.23\ \ 23\ \ 23\ \ldots &= 5 + \frac{23}{100} + \frac{23}{(100)^2} + \frac{23}{(100)^3} + \cdots \\
                         &= 5 + \frac{23}{100} \underbrace{\left(1 + \frac{1}{100} + \left(\frac{1}{100}\right)^2 + \cdots \right)}_{1/(1-0.01)} \\
                         &= 5 + \frac{23}{100}\left(\frac{1}{0.99}\right) = 5 + \frac{23}{99} = \frac{518}{99}
\end{align}
