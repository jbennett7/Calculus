# Limits of sequences of numbers
Informally, a sequence is an ordered list of things, but in this chapter the things will usually be numbers. We have seen sequences before, such as the sequence $x_0, x_1,\ldots,x_n,\ldots$ of numbers generated by Newton's method and the sequence $c_1,c_2,\ldots,c_n,\ldots$ of polygons that define Helga von Koch's snowflake. These sequences have limits, but many equally important sequences do not.
