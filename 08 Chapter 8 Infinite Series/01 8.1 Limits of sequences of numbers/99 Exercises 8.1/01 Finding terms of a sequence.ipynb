# Finding terms of a sequence
Each of Exercises 1&#151;6 gives a formula for the $n$th term $a_n$ of a sequence $\{a_n\}$. Find the values of $a_1$, $a_2$, $a_3$, and $a_4$.

## 1.
$
\begin{align}
a_n=\frac{1-n}{n^2}
\end{align}
$

### Solution

$
\begin{align}
a_n=\frac{1-n}{n^2} \rightarrow a_1=0\text{, }\ \ \ a_2=-\frac{1}{4}\text{, }\ \ \ a_3=-\frac{2}{9}\text{, }\ \ \ a_4=-\frac{3}{16}\text{.}
\end{align}
$

## 2.

$
\begin{align}
a_n=\frac{1}{n!}
\end{align}
$

### Solution
$
\begin{align}
a_n=\frac{1}{n!} \rightarrow a_1=1\text{, }\ \ \ a_2=\frac{1}{2}\text{, }\ \ \ a_3=\frac{1}{6}\text{, } \ \ \ a_4=\frac{1}{24}\text{.}
\end{align}
$

## 3.
$
\begin{align}
a_n=\frac{(-1)^{n+1}}{2n-1}
\end{align}
$
### Solution
$
\begin{align}
a_n=\frac{(-1)^{n+1}}{2n-1} \rightarrow a_1=1\text{, }\ \ \ a_2=-\frac{1}{3}\text{, }\ \ \ a_3=\frac{1}{5}\text{, }\ \ \ a_4=-\frac{1}{7}\text{.}
\end{align}
$

## 4.
$
\begin{align}
a_n=2+(-1)^n
\end{align}
$
### Solution
$
\begin{align}
a_n=2+(-1)^n \rightarrow a_1=1\text{, }\ \ \ a_2=3\text{, }\ \ \ a_3=1\text{, }\ \ \ a_4=3\text{.}
\end{align}
$

## 5.
$
\begin{align}
a_n=\frac{2^n}{2^{n+1}}
\end{align}
$
### Solution
$
\begin{align}
a_n=\frac{2^n}{2^{n+1}} \rightarrow a_1=\frac{1}{2}\text{, }\ \ \ a_2=\frac{1}{2}\text{, }\ \ \ a_3=\frac{1}{2}\text{, }\ \ \ a_4=\frac{1}{2}\text{.}
\end{align}
$

## 6.
$
\begin{align}
a_n=\frac{2^n-1}{2^n}
\end{align}
$
### Solution
$
\begin{align}
a_n=\frac{2^n-1}{2^n} \rightarrow a_1=\frac{1}{2}\text{, }\ \ \ a_2=\frac{3}{4}\text{, }\ \ \ a_3=\frac{7}{8}\text{, }\ \ \ a_4=\frac{15}{16}\text{.}
\end{align}
$

Each of exercises 7&#151;12 gives the first term or two of a sequence along with a recursion formula for the remaining terms. Write out the first ten terms of the sequence.

## 7.
$
\begin{align}

\end{align}

\frac{31}{16}\text{, } \frac{63}{32}\text{, } \frac{127}{64}\text{, } \frac{255}{128}\text{, } \frac{511}{256}\text{, } \frac{1023}{512}\text{.}
\end{align}
$

## 11.

\begin{align}

\end{align}
$

$

## 12.
$
\begin{align}

\end{align}
$

$
\begin{align}

-1\text{, } -2\text{, } 2\text{, } -1\text{, } -\frac{1}{2}\text{, } \frac{1}{2}\text{.}
