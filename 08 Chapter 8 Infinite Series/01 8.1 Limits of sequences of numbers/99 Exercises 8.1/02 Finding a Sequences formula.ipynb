# Finding a sequence's formula
In Exercies 13&#151;22, find a formula for the $n$th term of the sequence.

## 13.
The sequence $1$, $-1$, $1$, $-1$, $1$, $\ldots$
### Solution
$
\begin{align}
a_n=(-1)^{n+1}
\end{align}
$

## 14.
The sequence $-1$, $1$, $-1$, $1$, $-1$, $\ldots$
### Solution
$
\begin{align}
a_n=(-1)^n
\end{align}
$

## 15.
The sequence $1$, $-4$, $9$, $-16$, $25$, $\ldots$
### Solution
$
\begin{align}
a_n=(-1)^{n+1}n^2
\end{align}
$

## 16.
The sequence $1$, $-\frac{1}{4}$, $\frac{1}{9}$, $-\frac{1}{16}$, $\frac{1}{25}$, $\ldots$
### Solution
$
\begin{align}
a_n=(-1)^{n+1}\frac{1}{n^2}
\end{align}
$

## 17.
The sequence $0$, $3$, $8$, $15$, $24$, $\ldots$
### Solution
$
\begin{align}
a_n=n^2-1
\end{align}
$

## 18.
The sequence $-3$, $-2$, $-1$, $0$, $1$, $\ldots$
### Solution
$
\begin{align}
a_n=-3+n
\end{align}
$

## 19.
The sequence $1$, $5$, $9$, $13$, $17$, $\ldots$
### Solution
$
\begin{align}
a_n=2n-1
\end{align}
$

## 20.
The sequence $2$, $6$, $10$, $14$, $18$, $\ldots$
### Solution

\end{align}
$

## 21.
The sequence $1$, $0$, $1$, $0$, $1$, $\ldots$
### Solution
$
\begin{align}
a_n=\frac{1+(-1)^{n+1}}{2}
\end{align}
$
