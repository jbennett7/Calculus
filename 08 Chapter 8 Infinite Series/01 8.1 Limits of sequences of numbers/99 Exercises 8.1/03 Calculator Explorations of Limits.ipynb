# Calculator Explorations of Limits

In Exercises 23&ndash;26, experiment with a calculator to find a value of $N$ that will make the inequality hold for all $n>N$. Assuming that the inequality is the one from the formal definition of the limit of a sequence, what sequence, is being considered in each case and what is its limit?

# 23
$
\begin{equation}
\left|\sqrt[n]{0.5}-1\right| < 10^{-3}
\end{equation}
$

```r
ex23 <- function(n){ abs(0.5^(1/n) - 1) }
condition <- 10^-3
n<-seq(1,1000,1)
ans23 <- 0
for(i in n){
    if(ex23(i)<condition){
        ans23 <- i
        break
    }
}
cat (ans23, "value of N for all n > N\n")
rm(ans23,condition,ex23,i,n)
```
